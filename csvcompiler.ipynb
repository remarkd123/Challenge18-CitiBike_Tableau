```python
# 3. Convert the 'tripduration' column to datetime datatype.
new_tripduration19 = citibike19_df.tripduration
new_tripduration20 = citibike20_df.tripduration
new_tripduration21 = citibike21_df.tripduration
new_tripduration22 = citibike22_df.tripduration

# print(new_tripduration)

new_tripduration_datetime19 = pd.to_datetime(new_tripduration19, unit='m')
new_tripduration_datetime20 = pd.to_datetime(new_tripduration20, unit='m')
new_tripduration_datetime21 = pd.to_datetime(new_tripduration21, unit='m')
new_tripduration_datetime22 = pd.to_datetime(new_tripduration22, unit='m')

# print(new_tripduration_datetime)

# Add a new column for the converted trip duration
citibike19_df['new_tripduration'] = new_tripduration_datetime19
citibike20_df['new_tripduration'] = new_tripduration_datetime20
citibike21_df['new_tripduration'] = new_tripduration_datetime21
citibike22_df['new_tripduration'] = new_tripduration_datetime22
```

```python
with open(os.path.join(Out_Dir, 'Output-summers.csv'), 'w', newline='') as f:
    wf = csv.writer(f, lineterminator='\n')

    for files in csv_file_list:
        with open(files, 'r') as r: 
            next(r)                   # SKIP HEADERS
            rr = csv.reader(r)
            for row in rr:
                wf.writerow(row)
```
